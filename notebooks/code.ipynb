```python
#Importing the OpenAI library and other necessary modules
# The OpenAI library is used to interact with the OpenAI API, while os and sys are used for handling environment variables and system-specific parameters. The dotenv library is used to load environment variables from a .env file.

from openai import OpenAI
import os 
import sys
from dotenv import load_dotenv


load_dotenv()

client = OpenAI(
    base_url="https://integrate.api.nvidia.com/v1",
    api_key=os.getenv("api_key")
)
```

```python
# This function will handle the main loop of the application, allowing the user to input questions and receive responses from the assistant until they choose to exit.
#We will be using mistralai/magistral-small-2506 as the model for generating responses. The function will also handle streaming responses, allowing the assistant to provide answers in real-time as they are generated.
def main_twp():  
    while True:
        # 1. Ask for input INSIDE the loop so it repeats
        # Fixed the variable name so it's consistent
        user_input = input("\n\nEnter your question (or 'exit' to quit): ")

        # 2. Check for the exit command immediately
        if user_input.lower() == "exit":
            print("Session ended.")
            break

        # 3. Make the API request
        try:
            completion = client.chat.completions.create(
                model="mistralai/magistral-small-2506",
                messages=[{"role": "user", "content": user_input}],
                temperature=0.7,
                max_tokens=200,
                stream=True
            )

            # 4. Process and print the stream
            print("Assistant: ", end="", flush=True)
            for chunk in completion:
                if chunk.choices:
                    content = chunk.choices[0].delta.content
                    if content:
                        print(content, end="", flush=True)
            print() # Adds a newline after the full response is done

        except Exception as e:
            print(f"\nSomething went wrong: {e}")

# 5. CRITICAL: You must call the function to start it
main_twp()
```

```python
#Now we will use the Kimi-K2 Thinking Model

def ask_once():
    user_input = input("Enter your question: ")

    if user_input.strip().lower() in ["exit", "quit"]:
        print("Session ended.")
        return

    try:
        completion = client.chat.completions.create(
            model="moonshotai/kimi-k2-thinking-model-2024-06",
            messages=[{"role": "user", "content": user_input}],
            temperature=1,
            top_p=0.9,
            max_tokens=200,
            stream=True
        )

        print("Assistant: ", end="", flush=True)
        for chunk in completion:
            if chunk.choices:
                content = chunk.choices[0].delta.content
                if content:
                    print(content, end="", flush=True)
        print()

    except Exception as e:
        print(f"\nError: {e}")

ask_once()
```

```python
from openai import OpenAI

client = OpenAI(
  base_url = "https://integrate.api.nvidia.com/v1",
  api_key = os.getenv("api_key")
)


completion = client.chat.completions.create(
  model="deepseek-ai/deepseek-v4-flash",
  messages=[{"role":"user","content":""}],
  temperature=1,
  top_p=0.95,
  max_tokens=16384,
  extra_body={"chat_template_kwargs":{"thinking":True,"reasoning_effort":"high"}},
  stream=True
)

for chunk in completion:
  if not getattr(chunk, "choices", None):
    continue
  reasoning = getattr(chunk.choices[0].delta, "reasoning", None) or getattr(chunk.choices[0].delta, "reasoning_content", None)
  if reasoning:
    print(reasoning, end="")
  if chunk.choices and chunk.choices[0].delta.content is not None:
    print(chunk.choices[0].delta.content, end="")
  

```
